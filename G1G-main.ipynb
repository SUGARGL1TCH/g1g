```python
#Initialisation
import ipywidgets as widgets
from IPython.display import display, clear_output, Image, Audio
import urllib.request
import time

baseURL="https://www.mederp.net/g1g"

#Data structures
room = {"name": "room","background": f"{baseURL}/scenes/room.png"}
house = {"name": "house","background": f"{baseURL}/scenes/house.png"}
cafe = {"name": "cafe","background": f"{baseURL}/scenes/cafe.png"}
cafeteria = {"name": "cafeteria","background": f"{baseURL}/scenes/cafeteria.png"}
street = {"name": "street","background": f"{baseURL}/scenes/street.png"}
classroom = {"name": "classroom","background": f"{baseURL}/scenes/classroom.png"}

scenes = {"room":room,"house":house,"street":street,"cafe":cafe,"cafeteria":cafeteria,"classroom":classroom}

scores = {
    "affection": 20,  # Starting value
    "health": 20,   # Starting value
    "money": 20      # Starting value (in dollars)
}

characters = {
    "narrator": {"name":"Narrator","state": "normal"},
    "mc": {"name":"MC","state": "normal"},
    "valenda": {"name":"Valenda","state": "normal"},
    "kadol": {"name":"Kadol","state": "normal"},
    "mother": {"name":"Mother","state": "normal"},
    "canaisa": {"name":"Canaisa","state": "normal"}
}

mc = input("What is the name of the main character?")
characters['mc']["name"] = mc

#Displays a character image based on the character dictionary. Example: character_dict = {"name": "Valenda", "state": "happy"}
def render_character(character_dict):
    name = character_dict["name"].lower()
    state = character_dict["state"].lower()
    image_url = f"{baseURL}/characters/{name}_{state}.png"

    try:
        # Create an Image widget
        img_widget = widgets.Image(value=open_url(image_url), format='png', width=200, height=300, layout=widgets.Layout(margin='5px'))
        display(img_widget)
    except:
        print(f"[{name} - {state} state]")

# Helper function to load image from URL
def open_url(url):
    #Opens a URL and returns the content as bytes. This is needed because IPython Image widget expects bytes.
    with urllib.request.urlopen(url) as response:
        return response.read()

def render_scene(scene_dict):
    # Displays a scene background image based on the scene dictionary. Example:{"name": "cafe", "background": "https://mederp.net/g1g/scenes/cafe.png"}
    background_url = scene_dict["background"]
    scene_name = scene_dict["name"]

    try:
        bg_widget = widgets.Image(value=open_url(background_url), format='png', width=800, height=600, layout=widgets.Layout( margin='0px', border='2px solid black'))

        # Add a small caption with the scene name
        print(f"📍 Location: {scene_name.capitalize()}")
        display(bg_widget)

    except Exception as e:
        # Fallback if image can't be loaded
        print(f"📍 Location: {scene_name.capitalize()} (background image not available)")
        print(f"   Error: {e}")


```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output

# --- Global Game State ---
current_stage = "stage_1"
current_dialog_index = 0

# --- UI Components ---
# We use Output widgets to prevent the screen from jumping when updating images
out_scene = widgets.Output()
out_char = widgets.Output()

# Text display for dialogs and status
html_dialog = widgets.HTML(value="")

# Navigation controls
btn_next = widgets.Button(description="Next >", button_style='info')
box_choices = widgets.VBox() # Container for dynamic choice buttons

# Layout composition
ui_images = widgets.HBox([out_scene, out_char])
ui_controls = widgets.VBox([html_dialog, btn_next, box_choices])
main_ui = widgets.VBox([ui_images, ui_controls])

# --- Core Engine Functions ---
def update_ui():
    """Updates the screen based on the current stage and dialog index."""
    global current_stage, current_dialog_index, scores

    # Check if we reached the end of the game
    if current_stage == "endings":
        show_endings()
        return

    stage_data = story[current_stage]
    scene_name = stage_data["scene"]
    dialogs = stage_data["dialogs"]

    # 1. Render Scene Background
    with out_scene:
        clear_output(wait=True)
        render_scene(scenes[scene_name])

    # 2. Handle Dialogs vs. Choices
    if current_dialog_index < len(dialogs):
        # We are still reading through dialogs
        current_dialog = dialogs[current_dialog_index]
        speaker_key = current_dialog["character"]

        # Update the character's state dynamically
        speaker = characters[speaker_key]
        speaker["state"] = current_dialog["state"]

        # Render the Character Image
        with out_char:
            clear_output(wait=True)
            render_character(speaker)

        # Display the formatted text
        html_dialog.value = f"<h3 style='margin:0;'>{speaker['name']}</h3><p style='font-size:16px;'>{current_dialog['text']}</p>"

        # Ensure 'Next' button is visible and choices are hidden
        btn_next.layout.display = 'block'
        box_choices.layout.display = 'none'
        box_choices.children = [] # Clear any previous buttons

    else:
        # Dialogs are finished; time to present choices
        html_dialog.value = "<h3 style='margin:0;'>Make a decision:</h3>"

        # Clear character image to focus on choices
        with out_char:
            clear_output(wait=True)

        btn_next.layout.display = 'none'
        box_choices.layout.display = 'block'

        # Generate a button for each choice
        choice_buttons = []
        for choice in stage_data["choices"]:
            btn = widgets.Button(
                description=choice["text"],
                layout=widgets.Layout(width='auto', margin='5px')
            )
            btn.choice_data = choice # Store the choice dictionary inside the button object
            btn.on_click(handle_choice)
            choice_buttons.append(btn)

        box_choices.children = choice_buttons

# --- Event Handlers ---
def handle_next(b):
    """Triggered when the 'Next' button is clicked."""
    global current_dialog_index
    current_dialog_index += 1
    update_ui()

# Bind the click event to the Next button
btn_next.on_click(handle_next)

def handle_choice(b):
    """Triggered when a choice button is clicked."""
    global current_stage, current_dialog_index, scores

    # Extract the data attached to the button
    choice = b.choice_data

    # Apply statistical modifiers to global scores
    scores["money"] += choice["money"]
    scores["affection"] += choice["affection"]
    scores["health"] += choice["health"]

    # Transition to the next stage
    current_stage = choice["next"]
    current_dialog_index = 0

    update_ui()

# --- Ending Calculation ---
def show_endings():
    """Calculates and displays the final ending based on scores."""
    # Clear the image outputs
    with out_scene:
        clear_output()
    with out_char:
        clear_output()

    btn_next.layout.display = 'none'
    box_choices.layout.display = 'none'

    # Evaluate conditions
    valid_endings = []
    if scores["money"] > 1000:
        valid_endings.append(("money", scores["money"], "Ending: Rich Business Owner! You started your own business and escaped poverty."))
    if scores["affection"] > 50:
        valid_endings.append(("affection", scores["affection"], "Ending: True Happiness! You live happily with family, friends, and met the love of your life."))
    if scores["health"] > 50:
        valid_endings.append(("health", scores["health"], "Ending: Inner Strength! You became a strong, reliable young lady and overcame your problems."))

    # Determine the highest valid score if multiple conditions are met
    if not valid_endings:
        final_text = "Game Over. You thought poverty was a game?!"
    else:
        # Sort by the score value (index 1 of the tuple) in descending order
        valid_endings.sort(key=lambda x: x[1], reverse=True)
        final_text = valid_endings[0][2]

    html_dialog.value = f"<h2>Conclusion</h2><p style='font-size:18px;'><b>{final_text}</b></p><hr/><p>Final Scores: {scores}</p>"

# --- Bootstrapper ---
def start_game():
    """Initializes and displays the game UI."""
    global current_stage, current_dialog_index
    current_stage = "stage_1"
    current_dialog_index = 0
    display(main_ui)
    update_ui()

# Start the game loop
start_game()
```
